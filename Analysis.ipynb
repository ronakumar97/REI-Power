```python
time_rows = 0

for index, row in df.iterrows():
    try:
        if(row['SQ1_CP1_DF_OAT'].startswith('TIME')):
            print(row['Time'], row['SQ1_CP1_DF_OAT'])
            time_rows += 1
    except:
        continue
```
